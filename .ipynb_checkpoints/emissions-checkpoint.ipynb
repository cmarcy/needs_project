```python
#importing packages needed for analysis
import os
import numpy as np
import pandas as pd
import math
from pandas import DataFrame
from itertools import product

path = os.getcwd()
parent = os.path.dirname(path)
print(path)
print(parent)
#os.mkdir(path+'\outputs')
```

# Import stuff

```python
needs = pd.read_csv('outputs/needs_final.csv', low_memory=False)
needs['Primary_Fuel'] = needs['Modeled_Fuels'].str.split(",",expand=True)[0]
print(needs.columns)

egraw = pd.read_csv('inputs/egrid2019_data-PLNT19.csv',header = 1,thousands=',', low_memory=False)
egraw['PLNGENAN'] = egraw['PLNGENAN'].str.replace(",", "")
egraw['PLNGENAN'] = pd.to_numeric(egraw['PLNGENAN'],errors='coerce')
egraw['PLNGENOZ'] = egraw['PLNGENOZ'].str.replace(",", "")
egraw['PLNGENOZ'] = pd.to_numeric(egraw['PLNGENOZ'],errors='coerce')
egraw.head(100).to_csv("outputs\\check.csv",index = False)
print(egraw.columns[0:100])
print(egraw.columns[100:])
#print(egraw.dtypes.to_string())

col = pd.read_csv('inputs/egrid2019_data-PLNT19.csv', header= None, low_memory=False)
col = col[0:2].T
col.columns = ['name','id']
print(col.to_string())
```

```python
#calculate average emissions by primary fuel type

emit = egraw[['PLPRMFL','PLNGENAN','PLNGENOZ','PLNOXAN','PLNOXOZ','PLSO2AN','PLCO2AN','PLCH4AN','PLN2OAN','PLCO2EQA']]
emit_avg = emit.groupby('PLPRMFL', as_index=False).sum()

fuel = pd.read_csv('inputs/fuel_lookup.csv').drop(columns=['Fuel_Description'])
#print(fuel.head())
emit_avg = pd.merge(fuel,emit_avg,on=['PLPRMFL'],how='left')

emit_avg['PLNOXRTA'] = emit_avg['PLNOXAN'] / emit_avg['PLNGENAN'] * 2000
emit_avg['PLNOXRTO'] = emit_avg['PLNOXOZ'] / emit_avg['PLNGENOZ'] * 2000
emit_avg['PLSO2RTA'] = emit_avg['PLSO2AN'] / emit_avg['PLNGENAN'] * 2000
emit_avg['PLCO2RTA'] = emit_avg['PLCO2AN'] / emit_avg['PLNGENAN'] * 2000
emit_avg['PLCH4RTA'] = emit_avg['PLCH4AN'] / emit_avg['PLNGENAN']
emit_avg['PLN2ORTA'] = emit_avg['PLN2OAN'] / emit_avg['PLNGENAN']
emit_avg['PLC2ERTA'] = emit_avg['PLCO2EQA'] / emit_avg['PLNGENAN'] * 2000
#emit_avg['PLHGRTA'] = emit_avg['PLHGAN'] / emit_avg['PLNGENAN'] 

emit_avg = emit_avg.drop(columns=['PLPRMFL','PLNGENAN','PLNGENOZ','PLNOXAN','PLNOXOZ','PLSO2AN','PLCO2AN','PLCH4AN','PLN2OAN','PLCO2EQA'])
emit_avg['avg'] = 1
print(emit_avg.head())
#print(emit_avg.columns)
#emit_avg.to_csv("outputs\\emit_avg.csv",index = False)
```

```python
#Merge emission data to NEEDS

egrid = egraw[['ORISPL','OPRNAME','LAT','LON','CAPFAC','NAMEPCAP','PLNGENAN','PLNOXRTA',
               'PLNOXRTO','PLSO2RTA','PLCO2RTA','PLCH4RTA','PLN2ORTA','PLC2ERTA']]
egrid = egrid.rename(columns={'ORISPL':'ORIS_Plant_Code'})
egrid['egrid'] = 1

df = pd.merge(needs,egrid,on=['ORIS_Plant_Code'],how='left')
matched = df[df['egrid']==1].copy()
missing = df[df['egrid']!=1].copy()
missing = missing.drop(columns=['PLNOXRTA','PLNOXRTO','PLSO2RTA','PLCO2RTA','PLCH4RTA','PLN2ORTA','PLC2ERTA'])
missing = pd.merge(missing,emit_avg,on=['Primary_Fuel'],how='left')
#print(len(missing))
#missing.to_csv("outputs\\check_egrid.csv",index = False)

df2 = matched.append(missing)
#print(len(matched), len(missing), len(df2))

county = pd.read_csv('inputs/county.csv', encoding='latin-1')
fip = county[['COUNTYFP','LATITUDE','LONGITUDE']]
fip = fip.rename(columns={'COUNTYFP':'County_Code','LATITUDE':'county_LAT','LONGITUDE':'county_LON'})
#print(fip.head())

df3 = pd.merge(df2,fip,on=['County_Code'],how='left')
#print(df3.dtypes.to_string())

df3.loc[df3['LAT'].isnull(),'LAT'] = df3['county_LAT']
df3.loc[df3['LON'].isnull(),'LON'] = df3['county_LON']
```
